```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# We can see that Country Code is a common field so we can merge on that field
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
# how -> decides the kind of join, either left, right or inner
# We decided with left join as we wanted to work on the left side.
```

```python
final_df.head()
```

```python
### To check datatypes
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
plt.pie(country_values, labels = country_names)
```

```python
# Pie Chart of top 3 countries that uses zomato
plt.pie(country_values[:3], labels = country_names[:3], autopct="%1.2f%%")
```

Observation - Zomato's maximum records or transactions are from India
After that USA and then UK

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
# We can convert this to a dataframe
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
# Renaming the 0 column
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observations -
1. When rating is between is between 4.5 to 4.9, the rating is excellent!
2. When rating is between 4.0 to 4.4, the rating is very good!
3. When rating is between 3.5 to 3.9, the rating is good!
4. When rating is between 2.5 to 3.4, the rating is average!
5. When rating is lower than that, the rating is poor!

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green','green'])
```

Observation:
1. Not rated counted is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
# Plotting as a count plot - frequency of it being in the dataframe
sns.countplot(x="Rating color", data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green','green'])
```
